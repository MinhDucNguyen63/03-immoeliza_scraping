```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver

immoweb = 'https://www.immoweb.be/en'

status_url = immoweb + '/status'
status_response = requests.get(status_url)

if status_response.status_code == 200:
    print('Success!')
else:
    print('Error: ' + str(status_response.status_code))

cookie_url = immoweb + '/cookie'
cookies = requests.get(cookie_url).cookies
print(cookies)

immoweb_sale = immoweb + '/search/house/for-sale?countries=BE&page=1&orderBy=relevance'

immoweb_sale_response = requests.get(immoweb_sale)

print(immoweb_sale_response.text)
```

```python
DRIVER_PATH = '/path/to/chromedriver'

# Set up Chrome WebDriver
driver = webdriver.Chrome(executable_path=DRIVER_PATH)

# Navigate to Hacker News
driver.get("https://news.ycombinator.com/")

# Retrieve the page source
html = driver.page_source

# Close the driver
driver.quit()

# Parse the HTML with BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Find all 'tr' elements with class 'athing' which contain the news titles
titles = soup.find_all('tr', class_='athing')

# Loop through each title and print it
for title in titles:
    # Find the <a> tag within the 'titleline' span inside a 'td' with class 'title'
    title_link = title.find('td', class_='title').find('span', class_='titleline').find('a')
    title_text = title_link.get_text()  # Extract the text of the title
    print(title_text)
```
