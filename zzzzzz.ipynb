```python
#2nd approach:
import requests
from bs4 import BeautifulSoup

house_1 = 'https://www.immoweb.be/en/classified/house/for-sale/leuven/3000/20216274'
headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36"
    }
house_1_response = requests.get(house_1, headers=headers)
print(house_1_response)

soup = BeautifulSoup(house_1_response.content, 'html.parser')

#Property
html = soup.find("meta", {"property": "og:url"}).get("content")
html_list = html.split("/")
Property_ID = html_list[-1]
print('Property_ID : ', Property_ID)

# Locality name
html = soup.find("meta", {"property": "og:url"}).get("content")
html_list = html.split("/")
locality = html_list[-3]
print('Locality :', locality)

#Postal code
html = soup.find("meta", {"property": "og:url"}).get("content")
html_list = html.split("/")
postal_code = html_list[-2]
print('Postal_code : ', postal_code)
```
